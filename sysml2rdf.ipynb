```python
import json
import os

from src.sysml2rdf.parser import SlimRdfParser
from src.sysml2rdf.parser import RdfParser
from src.sysml2rdf.api.element_fetcher import ElementFetcher
```

```python
use_cache = True
```

```python
project_id = "477e1397-b094-4b94-a097-2cf8218d3110"
commit_id = "8af343ec-71ab-47ec-918c-0718424b85a4"
```

```python
filepath = f"cache/{project_id}.json"
data = {}

if not os.path.isfile(filepath) or not use_cache:
    fetcher = ElementFetcher(project_id, commit_id)
    data = fetcher.fetch().purge_empty_fields().get_data()

    with open(filepath, "w") as fp:
        json.dump(data, fp)
else:
    print(f"Load {filepath} from cache")
    with open(filepath) as fp:
        data = json.load(fp)
```

```python
parser = SlimRdfParser(data)
res = parser.parse()


filepath = f"cache/sysml_model_{project_id}.ttl"
with open(filepath, "w") as f:
    f.write(res)
    print(f"Successfully created file {filepath}")
```
